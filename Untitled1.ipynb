```python
from transformers import BartTokenizer, BartForConditionalGeneration
```

```python
tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")
model = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")

```

```python
text = """
Artificial Intelligence is transforming industries such as healthcare,
education, banking, transportation, and agriculture. AI helps doctors
diagnose diseases, assists teachers in personalized learning, detects
bank fraud, improves crop production, and powers autonomous vehicles.
It has become one of the most important technologies in the modern world.
"""
```

```python
inputs = tokenizer(
    text,
    return_tensors="pt",
    max_length=1024,
    truncation=True
)

summary_ids = model.generate(
    inputs["input_ids"],
    num_beams=4,
    max_length=80,
    min_length=25,
    early_stopping=True
)

summary = tokenizer.decode(
    summary_ids[0],
    skip_special_tokens=True
)

print("Original Text:\n")
print(text)

print("\n" + "="*60)

print("\nGenerated Summary:\n")
print(summary)
```

```python
# Word Count
original_words = len(text.split())
summary_words = len(summary.split())

# Character Count
original_characters = len(text)
summary_characters = len(summary)

# Compression Percentage
compression = (1 - (summary_words / original_words)) * 100

print("="*50)
print("TEXT SUMMARIZATION USING BART")
print("="*50)

print("\nOriginal Text:\n")
print(text)

print("\nGenerated Summary:\n")
print(summary)

print("\nStatistics")
print("----------------------------")
print("Original Words      :", original_words)
print("Summary Words       :", summary_words)
print("Original Characters :", original_characters)
print("Summary Characters  :", summary_characters)
print("Compression Rate    :", round(compression,2), "%")

# Save Summary
with open("summary.txt", "w") as file:
    file.write(summary)

print("\nSummary saved as summary.txt")
```
